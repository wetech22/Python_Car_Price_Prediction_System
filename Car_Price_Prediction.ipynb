```python
# Keeping only rows where 'year' is numeric
car = car[car['year'].str.isnumeric()]
```

```python
# Converting 'year' column to integer type
car['year'] = car['year'].astype(int)
```

```python
# Removing rows where 'price' is 'Ask For Price'
car = car[car['Price'] != 'Ask For Price']
```

```python
# Removing commas and converting 'price' to integer type
car['Price'] = car['Price'].str.replace(',', '').astype(int)
```

```python
# Cleaning the 'kms_driven' column
car['kms_driven'] = car['kms_driven'].str.split().str.get(0).str.replace(',', '')
```

```python
# Filtering out non-numeric 'kms_driven' values
car = car[car['kms_driven'].str.isnumeric()]
```

```python
car['kms_driven'] = car['kms_driven'].astype(int)
```

```python
# Removing rows where 'fuel_type' is NaN
car = car[~car['fuel_type'].isna()]
```

```python
# Keeping only the first three words from the 'name' column
car['name'] = car['name'].str.split(' ').str.slice(0, 3).str.join(' ')
```

```python
# Resetting the index after cleaning, and dropping the old index
car = car.reset_index(drop=True)
```
